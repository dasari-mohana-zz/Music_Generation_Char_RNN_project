## Visualizing Training loss and accuracy over number of Epochs

                                                                            Code Written By : Dasari Mohana

```python
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('logs/training_log.csv')
df
```

```python
plt.figure(figsize=(10, 10))
plt.plot(df['epoch'], df['loss'],c='r')
plt.title('Training Loss over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Training Loss')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 10))
plt.plot(df['epoch'], df['acc'],c='r')
plt.title('Training Accuracy over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Training Accuracy')
plt.grid()
plt.show()
```

# Conclusion
As number of epochs increases the training loss decreases and Accuracy increases
